```python
# core tensorflow modules
import tensorflow as tf
from tensorflow import keras

# using scikit for dataset and train_test_split apprently tf is difficult to work with in this case
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
# import some data to  with
db = datasets.load_boston()
X = db.data
Y = db.target

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=123)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(13,)),  # input layer
    keras.layers.Dense(64, activation=keras.activations.relu),  # hidden layer
    keras.layers.Dense(1, activation=keras.activations.relu) # output layer
])

model.compile(optimizer='adamax', loss=keras.losses.mean_absolute_error)
```

```python
model.fit(x_train, y_train, epochs=100) 
```

```python
test_loss = model.evaluate(x_test,  y_test, verbose=1) 

print('Test loss:', test_loss)
```

```python
# just to compare the loss values with an algorithm
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
m = ensemble.RandomForestRegressor()
m.fit(x_train, y_train)
mean_absolute_error(y_test, m.predict(x_test))
```
